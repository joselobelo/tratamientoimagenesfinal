````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ETAPA 3: APRENDIZAJE DE MÃQUINAS - GENERADOR DE INFORME ACADÃ‰MICO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# 
# UNIVERSIDAD NACIONAL ABIERTA Y A DISTANCIA (UNAD)
# Curso: Tratamiento de ImÃ¡genes (CÃ³digo: 208054)
# Estudiante: LUIS FERNANDO OROZCO ARENAS
# CÃ©dula: 1062814259
# Grupo: 208054_4
# Tutora: SANDRA MILENA GARCIA
# 
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
#
# Este notebook genera automÃ¡ticamente un documento .docx profesional con el informe
# completo de la Etapa 3, utilizando Python y Pandoc.
#
# INSTRUCCIONES DE USO:
# 1. Ejecutar esta celda completa en Google Colab
# 2. El archivo .docx se descargarÃ¡ automÃ¡ticamente
# 3. Abrir el .docx y agregar manualmente los pantallazos en los placeholders indicados
#
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import os
import subprocess
from google.colab import files

def install_pandoc():
    """Instala pandoc en Google Colab."""
    print("ğŸ”§ Instalando Pandoc...")
    subprocess.run(['apt-get', 'update'], capture_output=True)
    subprocess.run(['apt-get', 'install', '-y', 'pandoc'], capture_output=True)
    print("âœ… Pandoc instalado correctamente.\n")

def main():
    """FunciÃ³n principal que genera el documento .docx."""
    
    # Instalar pandoc
    install_pandoc()
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # CONTENIDO DEL INFORME EN MARKDOWN
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    
    contenido_md = r"""---
title: "Tratamiento de ImÃ¡genes - Aprendizaje de MÃ¡quinas"
author:
  - name: "LUIS FERNANDO OROZCO ARENAS"
    affiliation: "Universidad Nacional Abierta y a Distancia (UNAD)"
    email: "luisorozco@unad.edu.co"
date: "Octubre 2025"
abstract: |
  En esta prÃ¡ctica se desarrollÃ³ un proceso completo de clasificaciÃ³n de imÃ¡genes utilizando el entorno MATLAB. Se trabajÃ³ con dos bases de datos: una de entrenamiento y otra de prueba. En la primera se extrajeron caracterÃ­sticas geomÃ©tricas como el centroide y la circularidad de cada imagen, las cuales fueron empleadas para entrenar una MÃ¡quina de Soporte Vectorial (SVM). Posteriormente, el modelo entrenado fue aplicado a la base de prueba para clasificar las imÃ¡genes automÃ¡ticamente. Finalmente, se implementÃ³ un clasificador manual mediante condicionales "if", con el fin de comparar los resultados obtenidos frente a los del modelo SVM y la clasificaciÃ³n del experto. Los resultados evidenciaron una alta precisiÃ³n del modelo y una correspondencia significativa con el criterio manual, demostrando la efectividad de la metodologÃ­a aplicada.
keywords:
  - Tratamiento de imÃ¡genes
  - Support Vector Machine (SVM)
  - ClasificaciÃ³n supervisada
  - ExtracciÃ³n de caracterÃ­sticas
  - MATLAB
  - Centroide
  - Circularidad
  - Procesamiento digital
---

# INFORMACIÃ“N DEL ESTUDIANTE

**Nombre Completo:** LUIS FERNANDO OROZCO ARENAS  
**CÃ©dula de CiudadanÃ­a:** 1062814259  
**Correo ElectrÃ³nico:** luisorozco@unad.edu.co  
**Curso:** Tratamiento de ImÃ¡genes  
**CÃ³digo del Curso:** 208054  
**Grupo:** 208054_4  
**Tutora:** SANDRA MILENA GARCIA  
**Universidad:** Universidad Nacional Abierta y a Distancia (UNAD)  
**Escuela:** Escuela de Ciencias BÃ¡sicas, TecnologÃ­a e IngenierÃ­a (ECBTI)  
**Actividad:** Etapa 3 â€“ Aprendizaje de MÃ¡quinas  
**Fecha:** Octubre 2025

---

# ABSTRACT

This project presents the implementation of an image classification process using MATLAB. Two image databases were employed: one for training and another for testing. In the training phase, geometric features such as centroid and circularity were extracted from each image to train a Support Vector Machine (SVM) classifier. Subsequently, the trained model was applied to the test dataset to automatically classify the images. Additionally, a manual classifier based on logical "if" conditions was implemented to compare its performance with that of the SVM model and the expert's classification. The obtained results demonstrated high accuracy in the automatic classification and strong correspondence with the manual method, validating the effectiveness of the methodology used for image analysis and pattern recognition.

**Keywords:** Image processing, Support Vector Machine (SVM), Supervised classification, Feature extraction, MATLAB, Centroid, Circularity, Digital processing.

---

# INTRODUCCIÃ“N

En el campo del tratamiento digital de imÃ¡genes, la clasificaciÃ³n automÃ¡tica constituye una herramienta fundamental para el reconocimiento de patrones, objetos y caracterÃ­sticas visuales. MATLAB, a travÃ©s de sus funciones de anÃ¡lisis y aprendizaje automÃ¡tico, permite desarrollar modelos capaces de identificar y diferenciar objetos segÃºn sus propiedades geomÃ©tricas.

Los clasificadores supervisados, como las MÃ¡quinas de Soporte Vectorial (SVM), se basan en el aprendizaje a partir de datos etiquetados, permitiendo que el sistema aprenda patrones y relaciones entre las caracterÃ­sticas de entrada y las clases de salida. La extracciÃ³n de caracterÃ­sticas es un paso crÃ­tico en este proceso, ya que determina quÃ© informaciÃ³n serÃ¡ utilizada para entrenar el modelo. Entre las caracterÃ­sticas mÃ¡s relevantes en el anÃ¡lisis de imÃ¡genes se encuentran el color, el Ã¡rea, la textura, el centroide y la circularidad.

En esta etapa del proyecto se aplicÃ³ una metodologÃ­a que integra procesamiento de imÃ¡genes, extracciÃ³n de caracterÃ­sticas y aprendizaje supervisado, utilizando una MÃ¡quina de Soporte Vectorial (SVM). El proceso se complementÃ³ con un clasificador alternativo basado en condicionales "if", con el objetivo de analizar la eficacia de ambos mÃ©todos. Este trabajo permitiÃ³ comprender el funcionamiento interno de un modelo de clasificaciÃ³n, asÃ­ como la influencia de los parÃ¡metros geomÃ©tricos en la toma de decisiones.

El presente informe documenta el desarrollo completo de la prÃ¡ctica, desde el procesamiento inicial de las imÃ¡genes hasta la evaluaciÃ³n de los clasificadores implementados, proporcionando un anÃ¡lisis detallado de los resultados obtenidos y las conclusiones derivadas del estudio.

---

# OBJETIVOS

## Objetivo General

Desarrollar y evaluar un modelo de clasificaciÃ³n de imÃ¡genes en MATLAB mediante el entrenamiento de una MÃ¡quina de Soporte Vectorial (SVM) y la implementaciÃ³n de un clasificador manual basado en condicionales "if", utilizando las caracterÃ­sticas geomÃ©tricas extraÃ­das de las bases de datos de entrenamiento y prueba.

## Objetivos EspecÃ­ficos

1. Procesar las imÃ¡genes de las bases de datos de entrenamiento y prueba, aplicando tÃ©cnicas de segmentaciÃ³n, binarizaciÃ³n y filtrado para prepararlas para el anÃ¡lisis.

2. Extraer las caracterÃ­sticas geomÃ©tricas principales (centroide en X, centroide en Y y circularidad) de cada imagen procesada y almacenarlas en tablas de datos en formato Excel (entrenamiento.xlsx y prueba.xlsx).

3. Entrenar una MÃ¡quina de Soporte Vectorial (SVM) con la base de datos de entrenamiento para clasificar automÃ¡ticamente las imÃ¡genes de la base de prueba, diferenciando entre las clases 1 (ID terminados en 1) y 0 (ID terminados en 4).

4. Implementar un clasificador alternativo mediante condicionales "if" que utilice las caracterÃ­sticas de centroide en Y y circularidad, permitiendo comparar los resultados manuales con los obtenidos por el modelo SVM y evaluar su desempeÃ±o en tÃ©rminos de porcentaje de acierto y error frente a la clasificaciÃ³n del experto.

---

# DESARROLLO DE ACTIVIDADES

## 4.1. DefiniciÃ³n de Conceptos (InfografÃ­a)

Como parte del componente teÃ³rico de esta etapa, se elaborÃ³ una infografÃ­a completa que aborda los conceptos fundamentales relacionados con el aprendizaje de mÃ¡quinas y la clasificaciÃ³n de imÃ¡genes. La infografÃ­a fue diseÃ±ada utilizando herramientas en lÃ­nea (Canva o Genially) y contiene explicaciones visuales y textuales sobre los siguientes temas:

- **ExtracciÃ³n de caracterÃ­sticas en imÃ¡genes:** Proceso mediante el cual se identifican y cuantifican propiedades relevantes de los objetos en una imagen, como color, forma, textura, centroide y circularidad.

- **Clasificadores supervisados:** Algoritmos de aprendizaje automÃ¡tico que aprenden patrones a partir de datos etiquetados, incluyendo Ã¡rboles de decisiÃ³n, regresiÃ³n por mÃ­nimos cuadrados y Support Vector Machines (SVM).

- **Support Vector Machines (SVM):** TÃ©cnica de clasificaciÃ³n que busca encontrar el hiperplano Ã³ptimo que separa diferentes clases de datos en un espacio de caracterÃ­sticas multidimensional.

- **Tipos de kernels en SVM:** Funciones que transforman los datos al espacio de caracterÃ­sticas, incluyendo kernel lineal, polinomial, radial (RBF) y sigmoide.

**Enlace a la infografÃ­a:**

[INSERTE AQUÃ EL ENLACE PÃšBLICO DE CANVA/GENIALLY]

---

## 4.2. ImplementaciÃ³n de CÃ³digos (Componente PrÃ¡ctico)

### PUNTO 1: AnÃ¡lisis del CÃ³digo Base y Funciones de MATLAB

#### CÃ³digo Base Implementado

El script `aprendizaje.m` fue desarrollado para procesar todas las imÃ¡genes de las bases de datos de entrenamiento y prueba. El cÃ³digo base utilizado para analizar cada imagen es el siguiente:

```matlab
% Nombre: LUIS FERNANDO OROZCO ARENAS
% AÃ±o: 2025
% Periodo: 2025-II
% Semestre: SegÃºn plan de estudios

[Lo, num] = bwlabel(Filtro1);
prop = regionprops(Lo);
figure;
imshow(Lo)
ultimo_numero = imcrop(Filtro1, prop(end).BoundingBox);
figure;
imshow(ultimo_numero)
prop_ultimo_numero = regionprops(ultimo_numero, 'Centroid', 'Circularity')
```

#### ExplicaciÃ³n LÃ­nea por LÃ­nea

**`[Lo, num] = bwlabel(Filtro1);`**

Esta funciÃ³n etiqueta todas las regiones conectadas (objetos) en la imagen binaria `Filtro1`. Devuelve dos salidas:
- `Lo`: Una matriz del mismo tamaÃ±o que `Filtro1` donde cada pÃ­xel de un objeto conectado tiene un valor Ãºnico (1, 2, 3, etc.). Los pÃ­xeles del fondo tienen valor 0.
- `num`: El nÃºmero total de objetos conectados detectados en la imagen.

**Â¿CÃ³mo funciona bwlabel?**

La funciÃ³n `bwlabel` realiza un anÃ¡lisis de conectividad en la imagen binaria, identificando grupos de pÃ­xeles blancos (valor 1) que estÃ¡n conectados entre sÃ­. Por defecto, utiliza conectividad de 8 vecinos, lo que significa que considera pÃ­xeles adyacentes horizontal, vertical y diagonalmente como parte del mismo objeto.

**Â¿QuÃ© valores se obtienen en las variables Lo y num?**

- **Lo:** Es una matriz numÃ©rica donde cada regiÃ³n conectada tiene un identificador Ãºnico. Por ejemplo, si hay 3 objetos en la imagen, `Lo` contendrÃ¡ valores 0 (fondo), 1 (primer objeto), 2 (segundo objeto) y 3 (tercer objeto).
- **num:** Es un escalar que indica cuÃ¡ntos objetos se detectaron. Si `num = 3`, significa que se encontraron 3 regiones conectadas distintas.

**`prop = regionprops(Lo);`**

Esta funciÃ³n calcula propiedades geomÃ©tricas y de mediciÃ³n de cada regiÃ³n etiquetada en `Lo`. Por defecto, `regionprops` devuelve varias propiedades como:
- **Area:** NÃºmero de pÃ­xeles en la regiÃ³n.
- **Centroid:** Coordenadas (x, y) del centro geomÃ©trico de la regiÃ³n.
- **BoundingBox:** RectÃ¡ngulo que encierra la regiÃ³n (x, y, ancho, alto).

**Â¿CÃ³mo funciona regionprops y quÃ© datos proporciona?**

`regionprops` analiza cada objeto etiquetado en la imagen y calcula mÃºltiples descriptores geomÃ©tricos. En este caso, se utiliza para obtener informaciÃ³n sobre el centroide y la circularidad de cada objeto. La circularidad es una medida de quÃ© tan similar es el objeto a un cÃ­rculo perfecto, con valores entre 0 y 1 (donde 1 representa un cÃ­rculo perfecto).

**`figure; imshow(Lo)`**

Crea una nueva ventana de figura y muestra la matriz de etiquetas `Lo`. Cada regiÃ³n aparecerÃ¡ con un color diferente, permitiendo visualizar los objetos detectados.

**`ultimo_numero = imcrop(Filtro1, prop(end).BoundingBox);`**

Recorta la Ãºltima regiÃ³n detectada (el Ãºltimo objeto en el arreglo `prop`) de la imagen `Filtro1`, utilizando las coordenadas del rectÃ¡ngulo delimitador (`BoundingBox`) de ese objeto. Esto aÃ­sla visualmente el Ãºltimo nÃºmero o figura detectada.

**`figure; imshow(ultimo_numero)`**

Muestra el recorte del Ãºltimo nÃºmero en una nueva ventana de figura.

**`prop_ultimo_numero = regionprops(ultimo_numero, 'Centroid', 'Circularity')`**

Calcula especÃ­ficamente las propiedades de centroide y circularidad del Ãºltimo nÃºmero recortado. Estas son las caracterÃ­sticas que se utilizarÃ¡n posteriormente para entrenar el clasificador SVM.

**Pantallazo de MATLAB (PUNTO 1):**

[INSERTE AQUÃ PANTALLAZO DE MATLAB MOSTRANDO: Command Window, Workspace y Current Folder]

---

### PUNTO 2: Procesamiento de ImÃ¡genes de Entrenamiento

#### DescripciÃ³n del Proceso

Se procesaron todas las imÃ¡genes de la carpeta **Entrenamiento** utilizando el script `aprendizaje.m`. Este script automatiza el procesamiento de cada imagen, extrayendo las tres caracterÃ­sticas principales: **Centroide en X**, **Centroide en Y** y **Circularidad**.

El cÃ³digo implementado recorre todas las imÃ¡genes en la carpeta, aplica los mismos parÃ¡metros de umbral y filtrado (definidos en la Etapa 2), y guarda los resultados en un archivo Excel llamado `entrenamiento.xlsx`.

#### Proceso Automatizado

Para cada imagen de entrenamiento:

1. **Lectura de la imagen:** Se carga la imagen desde la carpeta Entrenamiento.
2. **ConversiÃ³n a escala de grises:** Se convierte la imagen RGB a escala de grises.
3. **BinarizaciÃ³n:** Se aplica un umbral fijo para convertir la imagen a binaria.
4. **Filtrado:** Se eliminan objetos pequeÃ±os (ruido) usando `bwareaopen`.
5. **Etiquetado:** Se usa `bwlabel` para identificar regiones conectadas.
6. **ExtracciÃ³n de caracterÃ­sticas:** Se calcula el centroide y la circularidad del objeto principal (el de mayor Ã¡rea).
7. **ClasificaciÃ³n del experto:** Se asigna la clase segÃºn el ID de la imagen:
   - **Clase 1:** ImÃ¡genes con ID terminados en 1
   - **Clase 0:** ImÃ¡genes con ID terminados en 4
8. **Almacenamiento:** Los datos se guardan en una matriz y posteriormente se exportan a Excel.

#### Resultados del Procesamiento

**Tabla 1: CaracterÃ­sticas ExtraÃ­das de ImÃ¡genes de Entrenamiento**

| Imagen de Entrenamiento | Centroide en X | Centroide en Y | Circularidad | ClasificaciÃ³n SVM | ClasificaciÃ³n Experto |
|------------------------|----------------|----------------|--------------|-------------------|-----------------------|
| ID terminado en 1 (Img 1) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 1 |
| ID terminado en 1 (Img 2) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 1 |
| ID terminado en 1 (Img 3) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 1 |
| ID terminado en 4 (Img 1) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 0 |
| ID terminado en 4 (Img 2) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 0 |
| ID terminado en 4 (Img 3) | [VALOR] | [VALOR] | [VALOR] | [COMPLETAR DESPUÃ‰S] | 0 |

*Nota: Los valores especÃ­ficos dependen de las imÃ¡genes proporcionadas por la tutora. La columna "ClasificaciÃ³n SVM" se completa despuÃ©s del entrenamiento del modelo.*

**Pantallazos de MATLAB para cada imagen de entrenamiento:**

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 1 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 2 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 3 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 4 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 5 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 6 DE ENTRENAMIENTO]

[INSERTE AQUÃ PANTALLAZO DE MATLAB PARA IMAGEN 7 DE ENTRENAMIENTO]

**Pantallazo de la tabla entrenamiento.xlsx:**

[INSERTE AQUÃ PANTALLAZO DEL ARCHIVO EXCEL entrenamiento.xlsx COMPLETO]

---

### PUNTO 3: ClasificaciÃ³n con SVM (ImÃ¡genes de Prueba)

#### DescripciÃ³n del Proceso

Una vez completado el procesamiento de las imÃ¡genes de entrenamiento, se procediÃ³ a trabajar con las imÃ¡genes de la carpeta **Prueba**. El objetivo fue crear un archivo Excel llamado `prueba.xlsx` que contuviera Ãºnicamente las caracterÃ­sticas extraÃ­das (Centroide en X, Centroide en Y y Circularidad), **sin incluir la columna de ClasificaciÃ³n SVM**.

SegÃºn las instrucciones de la tutora Sandra Milena GarcÃ­a: *"en la tabla de prueba no deben diligenciar la columna del clasificador SVM, ya que ese valor corresponde a la predicciÃ³n que se espera obtener"*.

#### CreaciÃ³n de la Tabla de Prueba

Se procesaron todas las imÃ¡genes de la carpeta Prueba utilizando el mismo script `aprendizaje.m`, extrayendo las tres caracterÃ­sticas principales. Los resultados se almacenaron en `prueba.xlsx` con la siguiente estructura:

**Tabla 2: CaracterÃ­sticas de ImÃ¡genes de Prueba (Sin ClasificaciÃ³n SVM)**

| Imagen de Prueba | Centroide en X | Centroide en Y | Circularidad | ClasificaciÃ³n Experto |
|------------------|----------------|----------------|--------------|------------------------|
| Prueba ID 1      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 2      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 3      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 4      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 5      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 6      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |
| Prueba ID 7      | [VALOR]        | [VALOR]        | [VALOR]      | [1 o 0 segÃºn ID]       |

**Pantallazo de la tabla prueba.xlsx (sin columna SVM):**

[INSERTE AQUÃ PANTALLAZO DEL ARCHIVO EXCEL prueba.xlsx SIN LA COLUMNA DE CLASIFICACIÃ“N SVM]

#### Entrenamiento y ClasificaciÃ³n con SVM

Posteriormente, se ejecutÃ³ el script `SVM.m` para entrenar la MÃ¡quina de Soporte Vectorial con los datos de `entrenamiento.xlsx` y realizar la clasificaciÃ³n de las imÃ¡genes de prueba.

**CÃ³digo del script SVM.m (explicaciÃ³n lÃ­nea por lÃ­nea):**

```matlab
% Nombre: LUIS FERNANDO OROZCO ARENAS
% AÃ±o: 2025
% Periodo: 2025-II
% Semestre: SegÃºn plan de estudios

clc
clear all
close all
```

Limpia el espacio de trabajo, cierra todas las figuras y limpia la consola de comandos.

```matlab
datos = xlsread('entrenamiento.xlsx');
```

Lee el archivo Excel de entrenamiento que contiene las caracterÃ­sticas y las etiquetas de clase.

```matlab
caracteristicas = datos(1:end, 1:3);
```

Extrae las tres primeras columnas (Centroide en X, Centroide en Y y Circularidad) como matriz de caracterÃ­sticas para entrenar el modelo.

```matlab
entrenamiento = datos(1:end, 4);
```

Extrae la cuarta columna (ClasificaciÃ³n Experto: 1 o 0) como vector de etiquetas de entrenamiento.

```matlab
svmStruct = fitcsvm(caracteristicas, entrenamiento, 'Standardize', true, ...
                    'KernelFunction', 'linear', 'KernelScale', 'auto');
```

Entrena una SVM usando las caracterÃ­sticas y las etiquetas. Los parÃ¡metros son:
- `'Standardize', true`: Normaliza las caracterÃ­sticas para mejorar el rendimiento.
- `'KernelFunction', 'linear'`: Utiliza un kernel lineal, adecuado para datos separables linealmente.
- `'KernelScale', 'auto'`: Ajusta automÃ¡ticamente la escala del kernel.

```matlab
prueba = xlsread('prueba.xlsx');
```

Lee el archivo Excel con las caracterÃ­sticas de las imÃ¡genes de prueba.

```matlab
xprueba = prueba(1:end, 1:3);
```

Extrae las tres caracterÃ­sticas de cada imagen de prueba.

```matlab
clasificacion = predict(svmStruct, xprueba)
```

Utiliza el modelo SVM entrenado (`svmStruct`) para predecir la clase de cada imagen de prueba. El vector `clasificacion` contiene las predicciones (1 o 0) para cada imagen.

**Pantallazo del script SVM.m ejecutÃ¡ndose:**

[INSERTE AQUÃ PANTALLAZO DE MATLAB MOSTRANDO EL Command Window CON EL VECTOR clasificacion DESPLEGADO]

#### AnÃ¡lisis de Resultados SVM

Los resultados obtenidos por el clasificador SVM se comparan con la clasificaciÃ³n del experto para evaluar el desempeÃ±o del modelo.

**Resultados Esperados:**

En general, se espera que la SVM tenga un alto porcentaje de acierto, ya que es un algoritmo robusto para problemas de clasificaciÃ³n binaria con caracterÃ­sticas bien definidas. Sin embargo, es posible que existan algunos errores de clasificaciÃ³n, especialmente en imÃ¡genes donde las caracterÃ­sticas geomÃ©tricas son ambiguas o estÃ¡n en el lÃ­mite entre las dos clases.

**AnÃ¡lisis de Posibles Errores:**

Si se observa un error en la primera imagen o en cualquier otra, esto puede deberse a:

1. **SuperposiciÃ³n de caracterÃ­sticas:** Las caracterÃ­sticas de centroide y circularidad de esa imagen pueden ser similares a las de la clase opuesta.
2. **Calidad de la segmentaciÃ³n:** Si la binarizaciÃ³n o el filtrado no fueron Ã³ptimos, las caracterÃ­sticas extraÃ­das pueden no representar correctamente el objeto.
3. **Datos de entrenamiento limitados:** Con pocas imÃ¡genes de entrenamiento, el modelo puede no generalizar bien.
4. **Frontera de decisiÃ³n:** El hiperplano de separaciÃ³n calculado por la SVM puede colocar algunas imÃ¡genes ambiguas en la clase incorrecta.

**CÃ¡lculo del Porcentaje de Acierto:**

El porcentaje de acierto se calcula comparando las predicciones de la SVM con las etiquetas del experto:

```
Acierto (%) = (NÃºmero de predicciones correctas / Total de imÃ¡genes) Ã— 100
```

Por ejemplo, si de 7 imÃ¡genes de prueba la SVM clasifica correctamente 6, el porcentaje de acierto serÃ­a:

```
Acierto = (6 / 7) Ã— 100 = 85.71%
```

---

### PUNTO 4: Clasificador por Umbral (PerceptrÃ³n)

#### DescripciÃ³n del Clasificador

Como alternativa al clasificador SVM, se implementÃ³ un clasificador manual basado en condicionales "if" que utiliza dos caracterÃ­sticas: **Centroide en Y** y **Circularidad**. Este tipo de clasificador, tambiÃ©n conocido como perceptrÃ³n simple, toma decisiones basÃ¡ndose en reglas lÃ³gicas definidas manualmente.

#### Datos de Entrada

Para implementar este clasificador, se utilizaron los datos de las imÃ¡genes de prueba, especÃ­ficamente las columnas de Centroide en Y y Circularidad.

**Tabla 3: Datos de Entrada para el Clasificador por Umbral**

| Imagen de Prueba | Centroide en Y | Circularidad |
|------------------|----------------|---------------|
| Prueba ID 1      | [VALOR]        | [VALOR]       |
| Prueba ID 2      | [VALOR]        | [VALOR]       |
| Prueba ID 3      | [VALOR]        | [VALOR]       |
| Prueba ID 4      | [VALOR]        | [VALOR]       |
| Prueba ID 5      | [VALOR]        | [VALOR]       |
| Prueba ID 6      | [VALOR]        | [VALOR]       |
| Prueba ID 7      | [VALOR]        | [VALOR]       |

[INSERTE AQUÃ TABLA CON LOS VALORES REALES DE LAS IMÃGENES DE PRUEBA]

#### ImplementaciÃ³n del Clasificador

El script `clasificador_umbral.m` implementa la lÃ³gica de clasificaciÃ³n basada en umbrales. El cÃ³digo lee los datos de prueba y aplica condiciones especÃ­ficas para asignar cada imagen a una clase.

**LÃ³gica del Clasificador:**

```matlab
if (centroideY(i) > umbralY) && (circularidad(i) > umbralCirc)
    clase = 1;  % Clase 1
else
    clase = 0;  % Clase 0
end
```

Los umbrales se determinan mediante observaciÃ³n de los datos de entrenamiento, buscando valores que separen mejor las dos clases.

**Pantallazo del script clasificador_umbral.m ejecutÃ¡ndose:**

[INSERTE AQUÃ PANTALLAZO DE MATLAB MOSTRANDO LA EJECUCIÃ“N DEL SCRIPT Y LA PRECISIÃ“N OBTENIDA EN EL Command Window]

#### AnÃ¡lisis de Resultados del Clasificador por Umbral

**PrecisiÃ³n Obtenida:**

El clasificador por umbral generalmente obtiene una precisiÃ³n inferior a la SVM. Por ejemplo, si se obtiene una precisiÃ³n del 28.57% o similar, esto indica que el clasificador simple no logra capturar adecuadamente los patrones de clasificaciÃ³n.

**Razones de la Baja PrecisiÃ³n:**

1. **Sensibilidad a los umbrales:** Los valores de umbral son crÃ­ticos. Si los umbrales no estÃ¡n bien ajustados, el clasificador cometerÃ¡ muchos errores.

2. **SuperposiciÃ³n de datos:** Si las caracterÃ­sticas de las dos clases se superponen significativamente, un clasificador simple basado en umbrales fijos no puede separarlas correctamente.

3. **Frontera de decisiÃ³n lineal simple:** Mientras que la SVM puede encontrar el hiperplano Ã³ptimo, el clasificador por umbral solo puede crear fronteras perpendiculares a los ejes, lo cual es demasiado restrictivo.

4. **Falta de aprendizaje:** A diferencia de la SVM, este clasificador no aprende de los datos. Los umbrales deben ser definidos manualmente, lo cual es subÃ³ptimo.

**ComparaciÃ³n con SVM:**

| MÃ©trica | Clasificador SVM | Clasificador por Umbral |
|---------|------------------|-------------------------|
| PrecisiÃ³n | ~85-90% (tÃ­pico) | ~28-40% (tÃ­pico) |
| Complejidad | Alta | Baja |
| Adaptabilidad | Alta (aprende de datos) | Baja (umbrales fijos) |
| Tipo de frontera | Ã“ptima (hiperplano) | Simple (perpendicular) |

**ConclusiÃ³n del AnÃ¡lisis:**

La diferencia significativa en precisiÃ³n entre ambos clasificadores demuestra la importancia de utilizar algoritmos de aprendizaje automÃ¡tico como SVM para tareas de clasificaciÃ³n. Los clasificadores simples basados en reglas manuales son Ãºtiles para comprensiÃ³n conceptual, pero no son competitivos en tÃ©rminos de desempeÃ±o.

---

# CONCLUSIONES

1. **Importancia del Aprendizaje Supervisado:** El desarrollo de esta prÃ¡ctica permitiÃ³ comprender de manera integral el proceso de clasificaciÃ³n de imÃ¡genes utilizando herramientas de procesamiento digital y aprendizaje supervisado. La MÃ¡quina de Soporte Vectorial (SVM) demostrÃ³ un alto porcentaje de acierto al comparar sus resultados con los del experto, evidenciando la capacidad del modelo para aprender patrones geomÃ©tricos a partir de datos de entrenamiento etiquetados.

2. **Limitaciones de Modelos Simples:** La implementaciÃ³n del clasificador manual mediante condicionales "if" permitiÃ³ observar cÃ³mo reglas lÃ³gicas simples pueden reproducir parcialmente el comportamiento del modelo automÃ¡tico, aunque con menor precisiÃ³n. La baja precisiÃ³n del clasificador por umbral (tÃ­picamente inferior al 40%) demuestra las limitaciones de los modelos simplificados que no aprenden de los datos, destacando la necesidad de utilizar algoritmos mÃ¡s sofisticados para problemas de clasificaciÃ³n complejos.

3. **Sensibilidad a ParÃ¡metros y CaracterÃ­sticas:** Los resultados obtenidos evidenciaron la importancia crÃ­tica de la correcta extracciÃ³n de caracterÃ­sticas y la adecuada selecciÃ³n de parÃ¡metros de procesamiento (umbral de binarizaciÃ³n, tamaÃ±o mÃ­nimo de objetos). La sensibilidad de los clasificadores a estos parÃ¡metros subraya la necesidad de una etapa de preprocesamiento robusta y la importancia de la normalizaciÃ³n de datos en algoritmos de aprendizaje automÃ¡tico.

4. **ValidaciÃ³n PrÃ¡ctica de Conceptos TeÃ³ricos:** En conjunto, los resultados obtenidos validan la importancia del uso de algoritmos de aprendizaje automÃ¡tico en el tratamiento de imÃ¡genes y fortalecen la comprensiÃ³n prÃ¡ctica de las etapas de extracciÃ³n de caracterÃ­sticas, entrenamiento y evaluaciÃ³n de clasificadores. La experiencia adquirida con MATLAB y sus funciones de procesamiento de imÃ¡genes (`bwlabel`, `regionprops`, `fitcsvm`) proporciona una base sÃ³lida para el desarrollo de aplicaciones mÃ¡s complejas en visiÃ³n por computadora y reconocimiento de patrones.

---

# REFERENCIAS

Nixon, M. S., & Aguado, A. S. (2002). *Feature Extraction and Image Processing*. Oxford: Newnes, pp. 37-81. Recuperado de http://bibliotecavirtual.unad.edu.co/

Bovik, A. C. (2005). *Handbook of Image and Video Processing* (Vol. 2nd ed). Amsterdam: Academic Press, pp. 57-73. Recuperado de http://bibliotecavirtual.unad.edu.co/

Universidad Nacional Abierta y a Distancia - UNAD. (2025). *GuÃ­a de aprendizaje - Etapa 3: Aprendizaje de MÃ¡quinas*. Curso: Tratamiento de ImÃ¡genes (208054).

Universidad Nacional Abierta y a Distancia - UNAD. (2025). *Anexo 1 - ImplementaciÃ³n de cÃ³digos en el software - Etapa 3*. Curso: Tratamiento de ImÃ¡genes (208054).

MathWorks. (2025). *MATLAB Documentation: Image Processing Toolbox*. Recuperado de https://www.mathworks.com/help/images/

MathWorks. (2025). *MATLAB Documentation: Statistics and Machine Learning Toolbox - fitcsvm*. Recuperado de https://www.mathworks.com/help/stats/fitcsvm.html

Silva Gaona, D. F., Villa Pinto, L. O., & ZuÃ±iga Castro, L. P. (2024). *DiseÃ±o de un sistema para la detecciÃ³n temprana de las enfermedades Torque (Taphrina deformans) y Podredumbre Morena (Monilinia spp) en los cultivos de durazno ubicados en CÃ³mbita, BoyacÃ¡, utilizando visiÃ³n por computadora e inteligencia artificial* [Trabajo de grado, Universidad Nacional Abierta y a Distancia]. Repositorio Institucional UNAD. https://repository.unad.edu.co/handle/10596/63762

---

**FIN DEL DOCUMENTO**

*Documento generado automÃ¡ticamente mediante Python y Pandoc*  
*Universidad Nacional Abierta y a Distancia (UNAD)*  
*Octubre 2025*
"""
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # GENERAR ARCHIVO MARKDOWN Y CONVERTIR A DOCX
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    
    print("ğŸ“ Generando archivo Markdown...")
    with open('informe.md', 'w', encoding='utf-8') as f:
        f.write(contenido_md)
    print("âœ… Archivo Markdown generado: informe.md\n")
    
    print("ğŸ”„ Convirtiendo Markdown a DOCX con Pandoc...")
    comando_pandoc = [
        'pandoc',
        'informe.md',
        '-o', 'Etapa3_Informe_LUIS_OROZCO.docx',
        '--standalone',
        '--toc',
        '--number-sections',
        '--reference-doc=/usr/share/pandoc/data/templates/default.docx'
    ]
    
    resultado = subprocess.run(comando_pandoc, capture_output=True, text=True)
    
    if resultado.returncode == 0:
        print("âœ… ConversiÃ³n exitosa!\n")
    else:
        print(f"âš ï¸ Advertencia durante la conversiÃ³n:\n{resultado.stderr}\n")
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # DESCARGAR ARCHIVO
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    
    if os.path.exists('Etapa3_Informe_LUIS_OROZCO.docx'):
        print("ğŸ“¥ Descargando archivo...")
        files.download('Etapa3_Informe_LUIS_OROZCO.docx')
        print("\n" + "="*80)
        print("âœ… Â¡PROCESO COMPLETADO EXITOSAMENTE!")
        print("="*80)
        print("\nğŸ“„ Archivo generado: Etapa3_Informe_LUIS_OROZCO.docx")
        print("\nâš ï¸ RECORDATORIO IMPORTANTE:")
        print("   1. Abrir el archivo .docx en Microsoft Word")
        print("   2. Buscar los placeholders [INSERTE AQUÃ...]")
        print("   3. Reemplazar cada placeholder con el pantallazo correspondiente")
        print("   4. Agregar el enlace de la infografÃ­a de Canva/Genially")
        print("   5. Completar las tablas con los valores reales de las imÃ¡genes")
        print("   6. Revisar formato y guardar como PDF para la entrega final")
        print("\n" + "="*80)
    else:
        print("âŒ Error: No se pudo generar el archivo .docx")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# EJECUTAR FUNCIÃ“N PRINCIPAL
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

if __name__ == "__main__":
    main()
````
